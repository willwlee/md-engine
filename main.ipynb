```python
from dynamics import NVE
import numpy as np
import matplotlib.pyplot as plt
from atoms import *

ts=0.01

dyn1 = NVE(x0=0, v0=5, mass=1, timestep=ts,algorithm="taylor")
traj1 = dyn1.run(1000)

dyn2 = NVE(x0=0, v0=5, mass=1, timestep=ts,algorithm="velocityverlet")
traj2 = dyn2.run(1000)

dyn3 = NVE(x0=0, v0=5, mass=1, timestep=ts,algorithm="leapfrogverlet")
traj3 = dyn3.run(1000)

x = np.linspace(0,10,1000)
plt.plot(x,traj1['E'], label="taylor")
plt.plot(x,traj2['E'], label="velocity")
plt.plot(x,traj3['E'], label="leapfrog")
plt.legend()
```
